```python
# Function for linear searching for a target
def linear_search(arr, target):
    # Loop through each index
    for i in range(0, len(arr)):
        # See if the item at the current index is equal to target
        if arr[i] == target:
            # Return the index if target found
            return i

    # Return -1 if target is not found    
    return -1
```

```python
arr = [3, 4, 1, 6, 9, 5]
linear_search(arr, 6)
```
