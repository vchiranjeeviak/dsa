```python
# Keep cards placing example in mind
# How you place cards in sorted order in one hand while taking each one from the pile
def insertion_sort(arr):
    # Going from 1st element to last leaving 0th
    for i in range(1, len(arr)):
        # Key is the current element
        key = arr[i]
        j = i - 1
        # Until previous index goes beyond 0 or key is greater than current
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        # Assign the key to appropriate index
        arr[j + 1] = key        
```

```python
arr = [2, 4, 7, 1, 9]
insertion_sort(arr)
arr
```
