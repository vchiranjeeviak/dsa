```python
def merge(arr, start, mid, end):
    # Getting lengths of left and right sub-arrays
    len_of_left = mid - start + 1
    len_of_right = end - mid
    
    # Creating left and right sub-arrays
    left = []
    right = []
    
    # Filling left and right sub-arrays
    for i in range(len_of_left):
        left.append(arr[start + i])
        
    for i in range(len_of_right):
        right.append(arr[mid + i + 1])
        
    index_of_left = 0
    index_of_right = 0
    # Considering only the part of the actual array which matters in this function call
    index_of_arr = start
    
    # Comparing both sub-arrays until one of them is empty
    while index_of_left < len_of_left and index_of_right < len_of_right:
        if left[index_of_left] <= right[index_of_right]:
            arr[index_of_arr] = left[index_of_left]
            index_of_left += 1
        else:
            arr[index_of_arr] = right[index_of_right]
            index_of_right += 1
        
        index_of_arr += 1
        
    # Fill up with the remaining one
    while index_of_left < len_of_left:
        arr[index_of_arr] = left[index_of_left]
        index_of_left += 1
        index_of_arr += 1
    
    while index_of_right < len_of_right:
        arr[index_of_arr] = right[index_of_right]
        index_of_right += 1
        index_of_arr += 1
```

```python
def merge_sort(arr, start, end):
    if start >= end:
        return
    
    mid = start + (end - start)//2
    merge_sort(arr, start, mid)
    merge_sort(arr, mid + 1, end)
    merge(arr, start, mid, end)
```

```python
arr = [1, 4, 1, 6, 8]
merge_sort(arr, 0, 4)
print(arr)
```
