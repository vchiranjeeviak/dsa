```python
def binary_addition_arrays(arr1, arr2):
    res = []
    len1 = len(arr1)
    len2 = len(arr2)

    if (len1 > len2):
        len1, len2 = len2, len1
        arr1, arr2 = arr2, arr1

    carry = 0
    for i in range(len2):
        sum = arr2[len2-i-1] + carry
        if len1 - i > 0:
            sum += arr1[len1-i-1]
        if sum > 1:
            carry = 1
            sum = sum % 2
        res.append(sum)
    res.append(carry)
        
    return res[::-1]
```

```python
arr1 = [1, 0, 1, 1]
arr2 = [1, 1, 1, 1]
binary_addition_arrays(arr1, arr2)
```
