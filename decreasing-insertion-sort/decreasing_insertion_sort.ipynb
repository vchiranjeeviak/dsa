```python
def decreasing_insertion_sort(arr):
    # Leave the first item and loop through remaining array
    for i in range(1, len(arr)):
        # Consider the current item
        key = arr[i]
        # Index of the previous item
        j = i - 1
        # Keep shifting items in left sub array to right until we find an item greater than or eqaul to key
        while j >= 0 and arr[j] < key:
            arr[j + 1] = arr[j]
            j -= 1
        # Assign the key at correct position
        arr[j + 1] = key 
```

```python
arr = [1, 4, 7, 2, 9]
decreasing_insertion_sort(arr)
arr
```
